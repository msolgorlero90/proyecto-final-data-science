**Descripción del proyecto**

Este proyecto analiza la discografía de Taylor Swift a través de un conjunto de datos que reúne características musicales de sus canciones obtenidas desde Spotify. El análisis busca comprender si existen patrones que caracterizan cada era —por ejemplo, diferencias en la energía, la valencia emocional o la bailabilidad— y cómo estos elementos podrían relacionarse con la recepción del público (medida a través de la variable de popularidad).

# **Data Wrangling: Limpieza y transformación de datos**

**Valores nulos**

````python
df.isnull().sum()

df.dropna(inplace=True)
````

**Transformaciones de Fechas/Nombres de variables/Valores**

````python
df.rename(columns={'name': 'song'}, inplace=True)

df['release_date'] = pd.to_datetime(df['release_date'])

# Transformar variables numéricas a porcentajes (0–100)
df['acousticness_index'] = df['acousticness'] * 100
df['energy_index'] = df['energy'] * 100
df['danceability_index'] = df['danceability'] * 100
df['valence_index'] = df['valence'] * 100

df[['acousticness_index', 'energy_index', 'danceability_index', 'valence_index']].head()
````

En esta etapa se realizaron transformaciones sobre el dataset original con el objetivo de facilitar el análisis posterior.
Se renombró la columna name a song para mayor claridad. También se eliminaron valores faltantes y se transformó la fecha de lanzamiento a formato de fecha. También se crearon nuevas variables que representan distintos índices musicales en escala de 0 a 100: acousticness_index, energy_index, danceability_index y valence_index.
Estas variables permiten comparar de forma más clara las características musicales de cada Era y facilitar la interpretación visual en los gráficos del análisis exploratorio.

# **EDA: Análisis Exploratorio de datos**

**Análisis univariado**

````python
# Primer gráfico: histograma de energía
plt.figure(figsize=(8,5))
sns.histplot(df['energy'], kde=True)
plt.title("Distribución de la energía de las canciones")
plt.xlabel("Energía")
plt.ylabel("Cantidad")
plt.show()

# Segundo gráfico: boxplot de popularidad por Era
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Era', y='popularity', palette='pastel')
plt.title("Popularidad de canciones por Era")
plt.xlabel("Era")
plt.ylabel("Popularidad (Spotify)")
plt.xticks(rotation=45)  # inclina las etiquetas para que se lean mejor
plt.show()
````

**Análisis bivariado**

````python
# 1. Relación entre energía y popularidad
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='energy', y='popularity', hue='Era', palette='tab10', s=80)
plt.title("Relación entre energía y popularidad", fontsize=13)
plt.xlabel("Energía")
plt.ylabel("Popularidad")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 2. Promedio de bailabilidad por Era
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Era', y='danceability', palette='crest')
plt.title("Promedio de bailabilidad por Era", fontsize=13)
plt.xlabel("Era")
plt.ylabel("Danceability (bailabilidad promedio)")
plt.xticks(rotation=45)
plt.show()
````

En el primer gráfico se observa una relación positiva entre la energía de las canciones y su popularidad: los temas con valores altos de energía tienden a obtener mejores posiciones en Spotify.
En el segundo gráfico, se aprecia que las Eras 1989, Reputation y Lover destacan por su mayor nivel de bailabilidad, mientras que Folklore y Evermore se caracterizan por un tono más acústico e introspectivo.

**Análisis multivariado**

````python
sns.scatterplot(data=df, x='energy', y='valence', hue='Era', size='popularity')
plt.title("Relación entre energía, valencia y popularidad por Era")


indices = ['acousticness_index', 'energy_index', 'danceability_index', 'valence_index']

df_melted = df.melt(
    id_vars=['Era'],
    value_vars=indices,
    var_name='Index',
    value_name='Value'
)

# Crear gráfico
plt.figure(figsize=(12,6))
sns.barplot(data=df_melted, x='Era', y='Value', hue='Index', palette='mako')

# Personalizar diseño
plt.title("Comparación de índices musicales por Era", fontsize=14, weight='bold')
plt.xlabel("Era")
plt.ylabel("Valor promedio del índice (0–100)")
plt.xticks(rotation=45)
plt.legend(title="Índice musical", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

Este gráfico permite comparar las características musicales promedio de cada Era de Taylor Swift a través de cuatro índices: acousticness, energy, danceability y valence.
Se observa que las Eras 1989 y Reputation presentan los niveles más altos de energía y bailabilidad, reflejando su perfil pop y orientado al ritmo. En contraste, Folklore y Evermore muestran los valores más altos de acousticness, lo que coincide con su estilo más introspectivo y narrativo.
Lover y Midnights exhiben un equilibrio entre energía y positividad emocional, mientras que The Tortured Poets Department se posiciona en un punto intermedio, con matices más melancólicos y suaves.

**Conclusiones del EDA**

El análisis exploratorio permitió comprender la evolución musical de Taylor Swift a lo largo de sus distintas Eras, observando cómo las métricas sonoras varían en función de los estilos y momentos artísticos de cada período.

````python
df = pd.read_csv("/Taylor Swift Database.csv")
df.head()
````

**Selección de Features**

````python
# Variables numéricas que vamos a usar como explicativas
features = ['acousticness', 'danceability', 'energy', 'valence',
            'tempo', 'speechiness', 'loudness']

# Matriz de features
X = df[features]

# Variable objetivo: popularidad
y = df['popularity']
````

````python
data = pd.concat([X, y], axis=1)
data = data.dropna()

X = data[features]
y = data['popularity']
````

````python
selector = SelectKBest(score_func=f_regression, k=3)
selector.fit(X, y)

selected_features = X.columns[selector.get_support()]
print("Variables seleccionadas:", list(selected_features))
````

**Preparación del Modelo**


````python
# 1. Armamos un dataframe solo con las columnas del modelo
model_df = df[['acousticness', 'energy', 'speechiness', 'popularity']].copy()

# 2. Eliminamos filas con valores faltantes en cualquiera de esas columnas
model_df = model_df.dropna()

# 3. Definimos X e y desde ESTE df limpio
X = model_df[['acousticness', 'energy', 'speechiness']]
y = model_df['popularity']

print(X.shape, y.shape)
````

````python
# Eliminar todas las filas con valores faltantes
df = df.dropna().reset_index(drop=True)
````

````python
from sklearn.feature_selection import SelectKBest, f_regression

# Asegurar que no haya valores faltantes
df = df.dropna().reset_index(drop=True)

# Features que seleccionaste
selected = ['acousticness', 'energy', 'speechiness']

X_selected = df[selected]
y = df['popularity']
````

**Entrenamiento del Modelo**

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.2, random_state=42
)

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
````

**Evaluación del Modelo**

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
````

Estos valores indican que el modelo tiene un desempeño limitado. El MAE sugiere que, en promedio, el modelo se desvía alrededor de 7 puntos de popularidad respecto de los valores reales.
